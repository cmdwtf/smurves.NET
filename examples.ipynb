# Usage examples for Smurves

The below test cases are intended to give an overview of Smurves' functionality and intended use.

## Example 1

* Set the number of curves that should be generated to 5
* Generate the curves along x-axis measurement points from 0.0 to 10.0
* Constrain the curves' y-axis values to the interval form 0.0 to 2.0
* Use 500 measurement points spaced equally along the x-axis interval
* Constrain how many times curves change their trajectory to at most once
* Let curves converge on the left-hand side in a y-axis value of 1.0

```python
#Import the necessary libraries for plotting
import time
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline

# Import the primary function from Smurves
from smurves import surgebinder

# Start a timer to measure the runtime
start_time = time.clock()

# Generate curves with the parameters requested
curves = surgebinder(n_curves = 5,
                     x_interval = [0.0, 10.0],
                     y_interval = [0.0, 2.0],
                     n_measure = 500,
                     direction_maximum = 1,
                     convergence_point = [0.0, 1.0])

# Print the runtime
stop_time = time.clock()
print('\nRuntime: %f seconds' % (stop_time - start_time))

# Set the figure size
f, ax = plt.subplots(figsize=(10, 5))
#ax.set(xscale="log")

# Plot each curve in the list of curves
for i in range(0, len(curves)):
    sns.tsplot(np.asarray(curves[i])[:, 1], np.asarray(curves[i])[:, 0], ax = ax)

# Specify the x-axis and y-axis intervals
x_interval = [0.0, 10.0]
y_interval = [0.0, 2.0]

# Limit the plots to the above intervals
plt.ylim(y_interval[0], y_interval[1])
plt.xlim(x_interval[0], x_interval[1])
```

## Example 3

* Set the number of curves that should be generated to 10
* Generate the curves along x-axis measurement points from 5.0 to 1.0
* Constrain the curves' y-axis values to the interval form 0.5 to 9.5
* Use 500 measurement points spaced in a log-scale along the x-axis
* Constrain how many times curves change their trajectory to at most thrice
* Let curves converge on the left-hand side in a y-axis value of 4.5
* Don't let the curves deviate on the y-axis before 5.1 on the x-axis

```python
#Import the necessary libraries for plotting
import time
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline

# Import the primary function from Smurves
from smurves import surgebinder

# Start a timer to measure the runtime
start_time = time.clock()

# Generate curves with the parameters requested
curves = surgebinder(n_curves = 10,
                     x_interval = [5.0, 10.0],
                     y_interval = [0.5, 9.5],
                     n_measure = 500,
                     direction_maximum = 3,
                     convergence_point = [5.0, 4.5],
                     log_scale = True,
                     start_force = 5.5)

# Print the runtime
stop_time = time.clock()
print('\nRuntime: %f seconds' % (stop_time - start_time))



# Set the figure size
f, ax = plt.subplots(figsize=(10, 5))

# Set the x-axis to the logarithmic scale
ax.set(xscale="log")

# Plot each curve in the list of curves
for i in range(0, len(curves)):
    sns.tsplot(np.asarray(curves[i])[:, 1], np.asarray(curves[i])[:, 0], ax = ax)

# Specify the x-axis and y-axis intervals
x_interval = [5.0, 10.0]
y_interval = [0.5, 9.5]

# Limit the plots to the above intervals
plt.ylim(y_interval[0], y_interval[1])
plt.xlim(x_interval[0], x_interval[1])
```

## Example 4

* Set the number of curves that should be generated to 20
* Generate the curves along x-axis measurement points from 0.01 to 10.0
* Constrain the curves' y-axis values to the interval form 0.0 to 1.0
* Use 500 measurement points spaced in a log-scale along the x-axis
* Constrain how many times curves change their trajectory to at most 5 times
* Don't set a curve convergence in a user-set point of origin for all curves
* Let the curves start their trajectories with randomly sampled angles

```python
#Import the necessary libraries for plotting
import time
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline

# Import the primary function from Smurves
from smurves import surgebinder

# Start a timer to measure the runtime
start_time = time.clock()

# Generate curves with the parameters requested
curves = surgebinder(n_curves = 20,
                     x_interval = [0.01, 10.0],
                     y_interval = [0.0, 1.0],
                     n_measure = 500,
                     direction_maximum = 5,
                     log_scale = True,
                     random_launch = True)

# Print the runtime
stop_time = time.clock()
print('\nRuntime: %f seconds' % (stop_time - start_time))



# Set the figure size
f, ax = plt.subplots(figsize=(10, 5))

# Set the x-axis to the logarithmic scale
ax.set(xscale="log")

# Plot each curve in the list of curves
for i in range(0, len(curves)):
    sns.tsplot(np.asarray(curves[i])[:, 1], np.asarray(curves[i])[:, 0], ax = ax)

# Specify the x-axis and y-axis intervals
x_interval = [0.01, 10.0]
y_interval = [0.0, 1.0]

# Limit the plots to the above intervals
plt.ylim(y_interval[0], y_interval[1])
plt.xlim(x_interval[0], x_interval[1])
```
